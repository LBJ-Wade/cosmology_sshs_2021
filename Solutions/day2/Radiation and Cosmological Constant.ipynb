
Recall that the first Friedmann equation is given by 
\begin{equation}
\left(\frac{da}{dt} a^{-1} \right)^2 = \frac{8 \pi G}{3 c^2} \mathcal{E}(t) - \frac{\kappa c^2}{R_0} \frac{1}{a^2(t)}
\end{equation}

where $a(t)$ is the value of the scale factor with time, $c$ is the speed of light, $R_0$ is the curvature radius of the Universe, $G$ is the gravitational constant, and $\mathcal{E}(t)$ is the energy density. 

# Problem 1

Make two separate plots, $d(t_{em})$ versus $z$ and $d(t_{obs})$ versus $z$, the proper distances to sources at the time that light was emitted and the proper distance to sources  versus $z$ today comparing a matter dominated flat universe, a radiation dominated flat universe, a cosmological constant dominated flat universe, and an empty negative curvature universe.

To answer this question, we are going to need $a(t)$ for both a radiation dominated and cosmological constant dominated Universe. Lets start with radiation dominated. We derived in class that
\begin{equation}
\mathcal{E}_w(a) \propto a^{-3(1+w)}
\end{equation}
For radiation, $w=1/3$

In both the matter and radiation dominated cases, setting $\dot{a}/a |_{t_0} = H_0$. 
\begin{equation}
\mathcal{E}_0 = \frac{3 c^2 H_0^2}{8 \pi G}
\end{equation}

For Dark Energy dominated
\begin{equation}
\left(\frac{\dot{a}}{a}\right)^2 = H_0^2
\end{equation}
Thus
\begin{equation}
\dot{a} = H_0 a
\end{equation}
which has the solution
\begin{equation}
a(t) = e^{H_0(t-t_0)}
\end{equation}
where $t_0$ is the present day. 

For radiation dominated
\begin{equation}
\dot{a} = H_0 a^{-2}
\end{equation}
Thus, if we set $a(t_0) = 1$ and $a(t=0)=0$,
\begin{equation}
a(t) = (2 t H_0)^{1/2}
\end{equation}

We showed before that in a vacuum dominated Universe, the proper distance is given by
\begin{equation}
d_p(t_0) = \frac{c}{H_0} z
\end{equation}
For a radiation dominated universe, which is representative of the conditions that existed right after the big bang, we have
\begin{equation}
d_p(t_0) = r = \int_{t_{em}}^{t_0} \frac{c dt}{a(t)} = \int_{t_{em}}^{t_0}  \frac{c dt}{\sqrt{t/t_0}} = c t_0 \left[1 - \sqrt{t_{em} / t_0} \right] = \frac{c}{2 H_0}\left[1 - \frac{1}{(1+z)}\right]
\end{equation}

For the rest of the Universes, we can use our answers from the matter dominated and curvature dominated cases.

# Problem 2

Plot the physical distance that light traveled between being emitted and being observed for the four Universes in Problem 1 as a function of $z$. How does it compare to $d(t_{em})$ and $d(t_{obs})$ for these four universes?



**Solution**
The physical distance that light travels is $c(t_0 - t_e)$. 

* Dark Energy Dominated: c($t_0 - t_e) = -cH_0^{-1} \ln a(t_e) = c H_0^{-1} \ln (1+z)$.
* Flat Matter Dominated: $c t_0 ( 1 - t_e/t_0) = \frac{2}{3} \frac{c}{H_0} \left[ 1 - \frac{1}{(1+z)^{3/2}} \right]$
* Flat Radiation Dominated: $c t_0 (1 - t_e / t_0) = \frac{c}{2 H_0} \left[ 1 - \frac{1}{(1+z)^2}\right]$
* Empty Curvature Dominated: $c t_0 (1 - t_e/t_0) = \frac{c}{H_0}\left[ 1 - \frac{1}{(1+z)}\right]$

Lets compare these to their proper distances.

```
fig, axarr = plt.subplots(1, 4)
plt.sca(axarr[0])
plt.plot(x, x-1, label='Proper Distance')
plt.plot(x, np.log(x), ls='--', label='Light Travel Length')
plt.plot(x, (x-1)/x, ls=':', label='Proper Distance at Emission')
plt.grid()
plt.title('Dark Energy Dominated Universe')
plt.xlabel('(1+z)')

plt.sca(axarr[1])
plt.plot(x,  2 * (1 - x ** (-.5)), label='Proper Distance')
plt.plot(x, 2./3. * (1 - 1/x**(1.5)), ls='--', label='Light Travel Length')
plt.plot(x,  2 * (1 - x ** (-.5))/x, ls=':', label='Proper Distance at Emission')
plt.grid()
plt.title('Flat Matter Dominated Universe')
plt.xlabel('(1+z)')

plt.sca(axarr[2])
plt.plot(x,  .5 * (1 - 1/x), label='Proper Distance')
plt.plot(x, .5 * (1 - 1/x**(2.)), ls='--', label='Light Travel Length')
plt.plot(x,  .5 * (1 - 1/x)/x, ls=':', label='Proper Distance at Emission')
plt.grid()
plt.title('Flat Radiation Dominated Universe')
plt.xlabel('(1+z)')

plt.sca(axarr[3])
plt.plot(x,  np.log(x), label='Proper Distance')
plt.plot(x, (1-1/x), ls='--', label='Light Travel Length')
plt.plot(x, np.log(x)/x, ls=':', label='Proper Distance at Emission')
plt.xlabel('(1+z)')
plt.grid()
plt.title('Empty Negative Curvature Universe')

plt.legend()
plt.gcf().set_size_inches(16, 4)
```

d_H(t) = a(t) \int_0^t  \frac{c dt^\prime}{a(t^\prime)}

Plot the comoving horizon for a matter dominated Universe.

# Problem 4

The comoving event horizon is the maximum comoving distance a particle can travel to an observer anytime in the future. The physical event horizon is the maximum physical distance at time $t$ that events can influence eachother in the future.

Plot the event horizon starting with today $t_0 H_0^{-1}$ to $10 t_0 H_0^{-1}$ if the Universe is flat and dark-energy dominated and assuming that $t_0 = H_0^{-1}$.

As the universe becomes infinitely old, what happens to the comoving event horizon?

**Solution**

The maximum comoving distance a partical can travel as $t \to \infty$ is given by 
\begin{equation}
r_\infty = \int_t^\infty \frac{c dt^\prime}{a(t^\prime)}
\end{equation}
For a dark-energy dominated universe we have
\begin{equation}
r_\infty = \int_t^\infty c dt e^{-H_0 (t - t_0)} = \frac{c}{H_0} e^{-H_0(t-t_0)}
\end{equation}
It follows here that the comoving horizon is actually shrinking exponentially as time increases! This means that galaxies in the hubble flow that we can observe today may not be observable in the far future! As the Universe becomes infinitely old, the comoving horizon shrinks to zero!

```
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 10, 100)
plt.plot(t, np.exp(-t))
plt.xlabel('$H_0 (t-t_0)$')
plt.ylabel('$r_\infty / (c / H_0)$')
plt.grid()
```
