# Problem 1

Roughly, what is the redshift of a source that emitted light on the day your were born but we are just observing now?

**Solution**
Suppose we observe emission from a source at proper distance $d_p \approx c \Delta t$ where $\Delta t$ is the time since your birthday. Using Hubble's law
\begin{equation}
(1+z) \approx \lambda_o / \lambda_e \approx 1 + v / c \approx 1+ H_0 d_p / c = 1 + H_0 \Delta t
\end{equation}
so the redshift of your birthday is approximately $H_0 \Delta t$. I was born roughly 32 years ago so my redshift is 
\begin{equation}
z \approx \frac{32}{14 \times 10^{9}} \approx 2 \times 10^{-9}
\end{equation}

# Problem 2

Suppose we observe a galaxy with redshift z and that the scale factor is evolving as $a(t) = e^{H_0(t-t_0)}$ where $H_0$ is the Hubble constant and is equal to 1/14 billion years and $t_0$ is the present day, $a(t_0) = 1$. Note that we are assuming an $a(t)$ here. This is not necessarily the $a(t)$ that describes our universe. We will learn later in the class how to derive $a(t)$ from first principals.




  Write down an expression for $\frac{dz_{obs}}{dt_{obs}}$ as a function of $H_0$ and $z_{obs}$?


  

**Solution**
Recall that the observed redshift of a sources is defined by
\begin{equation}
\frac{a(t_{obs})}{a(t_{em})} = 1 + z_{obs}
\end{equation}
If we take the derivative with respect to the observed time we get
\begin{equation}
\frac{d z_{obs}}{d t_{obs}} = \frac{d}{d t_{obs}} H_0 \frac{e^{H_0(t_{obs} - t_{0})}}{e^{H_0(t_{em} - t_{0})}} = \frac{H_0 e^{H_0(t_{obs} - t_0)}}{e^{H_0(t_{em} - t_0)}} - \frac{H_0 e^{H_0(t_{obs}-t_0)}}{e^{H_0(t_{em}-t_0)}} \frac{dt_{em}}{dt_{obs}}
\end{equation}
where we used the product rule and the chain rule for derivatives.

$dt_{em} / dt_{obs} = (1+z)^{-1}$ so we have
\begin{equation}
\frac{d z_{obs}}{d t_{obs}} = H_0(1+z_{obs}) - H_0\frac{(1+z)}{(1+z)} = H_0 z
\end{equation}

# Problem 3

Plot the proper distance of the Coma Cluster for the next ten billion years under this model, assuming that its current distance is 100 Mpc.

Plot the redshift of the Coma cluster for the next ten billion yaers assuming its current distance is 100 Mpc


**Solution**

The proper distance is $a(t) r$. If $a(t_0) = 1$, then $r=100$Mpc. We need to plot $100$Mpc $e^{H_0 t}$

```
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0,10, 100)
plt.plot(t, np.exp(t/14.) * 100.)
plt.xlabel('time from present [Gyr]')
plt.ylabel('$d_p$(t) [Mpc]')
plt.grid()
```

For redshift, use 
$dz / dt = H_0 z$. If we integrate this we get $z(t) = z_0 e^{H_0 t}$ where $z_0$ is the redshift today. To solve for the redshift today, we can use the fact that 
\begin{equation}
r = \int_{t_{em}}^{t_{obs}} \frac{c dt}{a(t)} = \frac{c}{H_0} \left[ e^{H_0(t_{obs} - t_0)} - e^{H_0(t_{em} - t_0)} \right] = \frac{c}{H_0} \left[ a(t_{obs}) - a(t_{em}) \right]
\end{equation}

Note that the last equality is not generally true, it only happens to work for our specific choice of $a(t)$! This means that 
\begin{equation}
\frac{H_0}{c}r = z_0
\end{equation}
so
\begin{equation}
z_0 = 100 \text{Mpc} / 4400 \text{Mpc} \approx 0.022
\end{equation}

```
t = np.linspace(0,10, 100)
plt.plot(t, np.exp(t/14.) * 0.022)
plt.xlabel('time from present [Gyr]')
plt.ylabel('redshift of Coma Cluster')
plt.grid()
```
