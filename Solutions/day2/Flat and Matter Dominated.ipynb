
Recall that the first Friedmann equation is given by 
\begin{equation}
\left(\frac{da}{dt} a^{-1} \right)^2 = \frac{8 \pi G}{3 c^2} \mathcal{E}(t) - \frac{\kappa c^2}{R_0^2} \frac{1}{a^2(t)}
\end{equation}

where $a(t)$ is the value of the scale factor with time, $c$ is the speed of light, $R_0$ is the curvature radius of the Universe, $G$ is the gravitational constant, and $\mathcal{E}(t)$ is the energy density. 

# Problem 1
 
Consider a universe that is completely dominated by matter. How would $\mathcal{E}(t)$ evolve with $a$. Recall that $\mathcal{E}(t) = \rho(t) c^2$ where $\rho(t)$ is the mass density and use the conservation of mass. 

**Solution**

For ordinary matter, $\mathcal{E}(t) \propto a^{-3}$. 

# Problem 2

Solve for $a(t)$ for a matter dominated universe with $\kappa = 0$ (flat, no curvature) if $a(t_0) = 1$  where $t_0=14$ billion years and $(\dot{a} / a(t_0)) = H_0$. 

**Solution**
With $\kappa = 0$, we have
\begin{equation}
\left(\frac{\dot{a}}{a} \right)^2 = \frac{8 \pi G}{3 c^2}\mathcal{E_0} a^{-3}
\end{equation}
We can solve for $\mathcal{E}_0$ using initial conditions, namely that 
\begin{equation}
H_0^2 = \frac{8 \pi G}{3 c^2} \mathcal{E}_0
\end{equation}
so that
\begin{equation}
\mathcal{E}_0 = \frac{3 H_0^2 c^2}{8 \pi G}
\end{equation}
Thus we have
\begin{equation}
\dot{a}^2 = H_0^2 a
\end{equation}
or
\begin{equation}
\dot{a} = H_0 a^{1/2}
\end{equation}
So
\begin{equation}
\int_0^{t} H_0 dt^\prime  = \int_{0}^a da^\prime (a^\prime)^{1/2}
\end{equation}
and
\begin{equation}
\frac{2}{3} a^{3/2} = H_0 t
\end{equation}
and
\begin{equation}
a(t) = \left(\frac{3}{2} H_0 t\right)^{2/3}
\end{equation}

# Problem 3
Plot the proper distance today to a source oberved with a redshift of $z$ as a function of $z$ in the matter dominated Universe you derived in problem 2. 

Compare this to the proper distance to a source at the time that the light was emitted as a function of redshift. 

**Solution**
Let $t_0 = (2/3 H_0^{-1})$. Since $a(t_0) = 1$, we have $a(t) = (t/t_0)^{2/3}$.

\begin{equation}
r = c \int_{t_e}^{t_o} \frac{dt}{a(t)} = \int_{t_e}^{t_0} \frac{c dt}{(t/t_0)^{2/3}} = 3 c t_0^{2/3} \left[ t_0^{1/3} - t_e^{1/3} \right] = 3ct_0 \left[ 1 - \left(\frac{t_e}{t_0}\right)^{1/3} \right] = 3c t_0 \left[ 1 - \sqrt{\frac{a(t_e)}{a(t_0)}} \right] = 3 c t_0 \left[ 1 - (1+z)^{-1/2}\right] = \frac{2 c}{H_0} \left[ 1- (1+z)^{-1/2}\right]
\end{equation}

Since $a(t_0)=1$, $r=d_p(t_0)$. For $d_p(t_e)$, just divide by $(1+z)$. 

```
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1, 10, 100)
plt.plot(x, 2 * (1 - x ** (-.5)), label='$d_p(t_0)$')
plt.plot(x, 2 * (1 - x ** (-.5)) / x, label='$d_p(t_e)$')

plt.ylabel('$d$ / $d_H$')
plt.xlabel('(1+z)')
plt.grid()
plt.legend()
```

# Problem 4

Do Problem 3 but now for sources in an empty open Univere with $\mathcal{E}(t) = 0$ and $\kappa = -1$.

Make a plot comparing the proper distance to a source at the time that light was emitted as a function of redshift comparing a universe with only curvature to a flat universe with only matter.


**Solution**

Now we have
\begin{equation}
\left(\frac{\dot{a}}{a}\right)^2 = \frac{c^2}{R_0^2}\frac{1}{a^2(t)}
\end{equation}
Evaluating our equation at $t=t_0$, we have
\begin{equation}
R_0 = \frac{c}{H_0}
\end{equation}
so
\begin{equation}
\dot{a} = H_0
\end{equation}
and
\begin{equation}
a(t) = H_0 t
\end{equation}

To compute $d_p$, we again take
\begin{equation}
r = \frac{c}{H_0} \int_{t_e}^{t_0} \frac{dt}{t} = \frac{c}{H_0} \ln\left(\frac{t_0}{t_e} \right) = \frac{c}{H_0} \ln\left(1+z\right)
\end{equation}

```
fig, axarr = plt.subplots(1, 2)
plt.sca(axarr[0])
plt.plot(x, np.log(x), label='$d_p(t_0)$, curvature dominated')
plt.plot(x, 2 * (1 - x ** (-.5)), label='$d_p(t_0)$ matter dominated')



plt.ylabel('$d$ / $d_H$')
plt.xlabel('(1+z)')
plt.grid()
plt.legend()
plt.sca(axarr[1])

plt.plot(x, np.log(x) / x, label='$d_p(t_e)$ curvature dominated')
plt.plot(x, 2 * (1 - x ** (-.5)) / x, label='$d_p(t_e)$ matter dominated')


plt.ylabel('$d$ / $d_H$')
plt.xlabel('(1+z)')
plt.grid()
plt.legend()

fig.set_size_inches(12, 6)
```

In a curvature dominated universe, sources at fixed redshift are further away then sources in a matter dominated universe.
