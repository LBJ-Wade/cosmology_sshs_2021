# Problem 1

# What are the metrics for the three different possible three-dimensional homogenous and isotropic spaces. 

**Solution**
* $ds^2 = dr^2 + R^2 \sin^2(r/R) [d \theta^2 + \sin^2(\theta) d\phi^2] $ for a positive curvature isotropic and homogenous space.

* $ds^2 = dr^2 + R^2[d \theta^2 + \sin^2(\theta) d \phi^2]$ for a zero curvature (flat) isotropic and homogenous space. 

* $ds^2 = dr^2 + R^2 \sinh^2(r/R) [d \theta^2 + \sin^2(\theta) d\phi^2] $ for a negative curvature isotropic and homegenous space.

# Problem 2

On the same plot, plot the circumference of a circle as a function of its radius $r$ (in units of Gigaparsecs)

* A 2d space with a positive radius of curvature of 10 Gpc
* A 2d space with a Negative radius of curvature of 5 Gpc
* A 2d flat space.

**Solution**

We can calculate the radius of a circule by considering a path in our two dimensional space at constant R with $dr=0$ so that $ds = d\theta$.

For a two dimensional flat space, we have

\begin{equation}
S = \int ds = r \int_0^{2 \pi} d \theta = 2 \pi r.
\end{equation}
 
 For a two dimensional positive radius space this is 
 \begin{equation}
 S = \int ds = 2 \pi R \sin(r/R)
 \end{equation}
 and for a negative curvature space this is
 \begin{equation}
S = 2 \pi R \sinh(r/R)
 \end{equation}

 Now lets plot S as a function or $r$.



```
import matplotlib.pyplot as plt
import numpy as np
r = np.linspace(0, 30, 100)
plt.plot(r, 2 * np.pi * 10. * np.sin(r / 10), label='positive curved space with 10 Gpc curvature radius')
plt.plot(r, 2 * np.pi * 5 * np.sinh(r / 5.), label='negative curved space with 5 Gpc curvature radius.')
plt.plot(r, 2 * np.pi * r, label='flat space')
plt.xlabel('r [Gpc]')
plt.ylabel('Circle Circumference [Gpc]')
plt.grid()
plt.legend()
plt.yscale('log')
```

# Problem 3
 (Ryden 3.2)

 Suppose you are a two-dimensional being, living on the surface of a sphere with radius R. A ladder with length $ds << R$ is at a distance $r$ from you and oriented perpindicular to your separation. 

* What angular width $d \theta$ will you measure for the object? Explain the behavior of $d\theta$ as $r \to \pi R$.
* Plot $d \theta$ versus $r/R$. 


**Solution**

We can describe the infinitesimal length $ds$ of an object as a small path length equal to 
\begin{equation}
ds = R \sin(r/R) d\theta
\end{equation}
Light arriving from either end of $ds$ will follow great circles, the straight lines on a spherical surface, from either end.
Thus
\begin{equation}
d \theta = \frac{ds}{R \sin(r/R)}
\end{equation}

As $r \to \pi R$, the angle $d \theta$ approaches infinity. This is because as the object approaches the sought pole of our sphere, the great circles intersecting its ends and the north pole have a larger and larger angle between them. When the object sits on the south pole, it will appear in all directions.

```
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, np.pi, 100)
plt.plot(x, 1./np.sin(x))
plt.xlim(0, np.pi)
plt.ylim(0, 10)
plt.xlabel('r/R')
plt.ylabel('d$\\theta$ / (ds / R)')
plt.grid()
```
